```python
base_url = "https://a2ch.ru/"
```

```python
pages =[]
```

```python
for year in range(2015, 2022):
    for month in range(1, 13):
        for date in range(1, 30):
            pages.append(f"{year}/{month}/{date}/")
```

```python
pages = pages[::-1]
pages = pages[70:]
```

```python
import requests 
from datetime import datetime
import pathlib
from bs4 import BeautifulSoup
import json
from tqdm.contrib.concurrent import process_map
import random
import time
```

```python
p = pathlib.Path(f"./arhive")
p.mkdir(parents=True, exist_ok=True)
```

```python
def save_thead(thread_url):
    page =  requests.get(base_url+thread_url)
    thread = BeautifulSoup(page.content)
    fn = f"{int(datetime.now().timestamp())}_{random.randint(0, 1000000)}.json"
    filepath = p / fn
    saved = []
    for post in thread.find_all(class_="post"):
        saved.append(post.find(class_="text").text)
    with filepath.open("w", encoding="utf-8") as f:
        json.dump(saved, f)
    time.sleep(0.1 + random.randint(10, 100)/10)
```

```python
def get_date_archive(url):
    page = requests.get(base_url+url)
    soup = BeautifulSoup(page.content)
    thred_urls = []
    for post in soup.find_all(class_="post"):
        thred_urls.append(post.find("a")["href"])
    process_map(save_thead, thred_urls, max_workers=6, chunksize=3)
```

```python
data = process_map(get_date_archive, pages[30+246:], max_workers=6, chunksize=3)
```
